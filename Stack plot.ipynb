```python
import matplotlib.pyplot as plt


```

```python
# Data
days = [1, 2, 3, 4, 5]
work = [8, 9, 8, 10, 7]
study = [2, 3, 2, 1, 3]
exercise = [1, 1, 1, 2, 1]
leisure = [3, 2, 3, 2, 3]
```

```python
# Creating the stack plot
plt.stackplot(days, work, study, exercise, leisure, 
              labels=['Work', 'Study', 'Exercise', 'Leisure'], 
              colors=['skyblue', 'lightgreen', 'orange', 'pink'])

# Adding titles and labels
plt.title('Stack Plot Example')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.legend(loc='upper left')

# Display the plot
plt.show()
```
