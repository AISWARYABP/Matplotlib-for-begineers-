```python
import matplotlib.pyplot as plt


```

```python
# Data
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]
sizes = [20, 50, 80, 40, 60, 70, 100, 30, 60, 90]  # Bubble sizes
```

```python
# Creating the scatter plot
plt.scatter(x, y, s=sizes, color='blue', edgecolor='black', alpha=0.8)

# Adding titles and labels
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```
