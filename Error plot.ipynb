```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Data
x = np.linspace(0, 10, 10)  # X-axis values
y = np.sin(x)               # Y-axis values (sin curve)
error = 0.1 + 0.2 * np.sqrt(x)  # Error values
```

```python
# Creating the error plot
plt.errorbar(x, y, yerr=error, fmt='o', ecolor='red', capsize=5, label='Data')

# Adding titles and labels
plt.title('Error Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the plot
plt.show()
```
