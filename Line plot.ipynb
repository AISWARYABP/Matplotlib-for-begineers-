```python
# Import necessary libraries
import matplotlib.pyplot as plt
```

```python
# Data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
```

```python
# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label="y = x^2")
# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot ")

# Show legend
plt.legend()

# Display the plot
plt.show()
```
