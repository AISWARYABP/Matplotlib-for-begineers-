```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate data
x = np.linspace(0, 10, 10)  # x values from 0 to 10
y = np.sin(x)               # y values as the sine of x
```

```python
# Create stem plot
plt.stem(x, y, linefmt="b-", markerfmt="bo", basefmt="r-", label="sin(x)")

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Stem Plot")

# Show legend
plt.legend()

# Display the plot
plt.show()
```
