```python
import matplotlib.pyplot as plt


```

```python
# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]
```

```python
# Creating the bar plot
plt.bar(categories, values, color='skyblue')

# Adding titles and labels
plt.title('Example Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Display the plot
plt.show()
```
