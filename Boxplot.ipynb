```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
data = [np.random.normal(50, 10, 200),  # Dataset 1: Mean=50, SD=10
        np.random.normal(60, 15, 200),  # Dataset 2: Mean=60, SD=15
        np.random.normal(70, 20, 200)]  # Dataset 3: Mean=70, SD=20
```

```python
# Creating the box plot
plt.boxplot(data, notch=True, patch_artist=True, 
            boxprops=dict(facecolor='skyblue', color='blue'))

# Adding titles and labels
plt.title('Box Plot Example')
plt.xlabel('Data Sets')
plt.ylabel('Values')

# Adding x-axis labels
plt.xticks([1, 2, 3], ['Dataset 1', 'Dataset 2', 'Dataset 3'])

# Display the plot
plt.show()
```
