```python
import matplotlib.pyplot as plt
import numpy as np


```

```python
# Data
data = [np.random.normal(60, 10, 100),  # Dataset 1: Mean=60, SD=10
        np.random.normal(70, 15, 100),  # Dataset 2: Mean=70, SD=15
        np.random.normal(80, 20, 100)]  # Dataset 3: Mean=80, SD=20
```

```python
# Creating the violin plot
plt.violinplot(data, showmeans=True, showextrema=True, showmedians=True)

# Adding titles and labels
plt.title('Violin Plot Example')
plt.xlabel('Dataset')
plt.ylabel('Values')

# Custom x-axis labels
plt.xticks([1, 2, 3], ['Dataset 1', 'Dataset 2', 'Dataset 3'])

# Display the plot
plt.show()
```
